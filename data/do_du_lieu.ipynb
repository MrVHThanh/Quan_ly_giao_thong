```python
import pandas as pd

# Đọc file
df_tuyen = pd.read_excel("giao_thong_data.xlsx", sheet_name="TuyenDuong")
df_doan = pd.read_excel("giao_thong_data.xlsx", sheet_name="DoanTuyen")
df_cap_ql = pd.read_excel("giao_thong_data.xlsx", sheet_name="CapQuanLy")
df_cap_duong = pd.read_excel("giao_thong_data.xlsx", sheet_name="CapDuong")
```

```python

# Rename bảng tuyến
df_tuyen = df_tuyen.rename(columns={
    'Mã tuyến': 'Ma_tuyen',
    'Cấp quản lý': 'Cap_quan_ly'
})
df_tuyen
```

```python
# Rename bảng đoạn
df_doan = df_doan.rename(columns={
    'Mã tuyến': 'Ma_tuyen',
    'Cấp đường': 'Cap_duong',
    'CD thực tế (km)': 'chieu_dai_thuc_te',
    'Lý trình đầu (km)': 'ly_trinh_dau',
    'Lý trình cuối (km)': 'ly_trinh_cuoi'
})
df_doan.head
```

```python
# JOIN
df = df_doan.merge(
    df_tuyen[['Ma_tuyen','Cap_quan_ly']],
    on='Ma_tuyen'
)
df
```

```python

df = df.merge(
    df_cap_ql.rename(columns={
        'ma_cap':'Cap_quan_ly',
        'ten_cap':'ten_cap_quan_ly',
        'thu_tu_hien_thi':'thu_tu_ql'
    }),
    on='Cap_quan_ly'
)
df
```

```python

df = df.merge(
    df_cap_duong.rename(columns={
        'Ma_cap':'Cap_duong',
        'ten_cap':'ten_cap_ky_thuat',
        'thu_tu_hien_thi':'thu_tu_cd'
    }),
    on='Cap_duong'
)
df.head
```

```python

# Tính chiều dài
df['chieu_dai'] = df['chieu_dai_thuc_te'].fillna(
    df['ly_trinh_cuoi'] - df['ly_trinh_dau']
)
```

```python

# Group
ketqua = df.groupby(
    ['Cap_quan_ly','ten_cap_quan_ly','thu_tu_ql',
     'Cap_duong','ten_cap_ky_thuat','thu_tu_cd'],
    as_index=False
)['chieu_dai'].sum()
ketqua
print(type(ketqua))
```

```python

# Sort
ketqua = ketqua.sort_values(['thu_tu_ql','thu_tu_cd'])

print(ketqua)
```
